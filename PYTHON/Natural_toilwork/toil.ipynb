```python
def summary1(summary):
    #upper-lower case change
    summary = summary.apply(lambda x: " ".join(x.lower() for x in x.split()))
    # poinst sign
    summary = summary.str.replace('[^\w\s]','')
    #numbers
    summary = summary.str.replace('\d','')
    #stopwords
    sw = stopwords.words("english")
    summary = summary.apply(lambda x: " ".join(x for x in x.split() if x not in sw ))
    ##lemmi
    from textblob import Word
    summary = summary.apply(lambda x : " ".join([Word(word).lemmatize() for word in x.split()]))
    summary = summary.str.replace("rt",'')
    freq_df = summary.apply(lambda x: pd.value_counts(x.split(" "))).sum(axis=0).reset_index()
    freq_df.columns = ["words", "frequency"]
    freq_df.groupby("words").count()["frequency"].sort_values(ascending=False)
    frequencies = freq_df[freq_df.frequency > freq_df.frequency.mean() + freq_df.frequency.std()] ##Reduction according to  std.

    return frequencies,summary
```

```python
def tfidf(data):
    tfidf = TfidfVectorizer(ngram_range=(2,2))
    sparse_matrix = tfidf.fit_transform(data)
    frequencies1 = sum(sparse_matrix).toarray()[0]
    df_grams = pd.DataFrame(frequencies1, index=tfidf.get_feature_names(), columns = ["frequency"])
    df_grams = df_grams.sort_values(by="frequency", ascending=False).head(30)
    return df_grams
```
