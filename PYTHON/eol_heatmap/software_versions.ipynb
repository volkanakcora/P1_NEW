```python
def data_processing_per_components(data_frame):
    """transforming data from row form to dataframe"""
    data_frame = data_frame.reset_index()
    data_frame = data_frame.rename(columns={'index': 'instance', 0: 'versions'})
    data_frame['per_components'] = data_frame['instance'][data_frame['instance'].str.contains(
        "per_component")]
    data_frame = data_frame[['versions', 'per_components']]
    data_frame['versions'] = data_frame['versions'].astype(str)  # convert to the str
    data_frame['versions'] = data_frame['versions'].map(lambda x: x.lstrip("[,'").rstrip("],'"))  # remove unwanted characters
    data_frame['versions'] = data_frame['versions'].map(lambda x: x.replace("'"," "))  # remove unwanted characters
    data_frame = data_frame.dropna()
    data_frame = data_frame.reset_index()
    data_frame = data_frame.drop(['index'], axis=1)
    data_frame["Timestamp"] = datetime.datetime.now()
    return data_frame
```

```python
df = read_json(path_to_json='/home/oh856/P1-REPO/PYTHON/eol_heatmap/data/')
df = data_processing_per_components(df)
```
