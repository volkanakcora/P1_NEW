```python
import math

def get_loan_principal():
    return int(input('Enter the loan principal: '))

def get_calculation_type():
    return input("""What do you want to calculate?
Type 'm' for the number of monthly payments,
Type 'p' for the monthly payment: """)

def calculate_months_to_repay_loan(loan_principal, monthly_payment):
    result = math.ceil(loan_principal / monthly_payment)
    return result

def calculate_monthly_payment(loan_principal, number_of_months):
    monthly_payment = loan_principal / number_of_months
    rounded_payment = math.ceil(monthly_payment)
    last_payment = loan_principal - (number_of_months - 1) * rounded_payment
    return rounded_payment, last_payment

def print_months_to_repay(result):
    print("It will take", result, "months to repay the loan.")

def print_monthly_payment(rounded_payment, last_payment):
    print("Your monthly payment =", rounded_payment, "and the last payment =", last_payment)

def main():
    loan_principal = get_loan_principal()
    action = get_calculation_type()

    if action == 'm':
        monthly_payment = int(input("Enter the monthly payment: "))
        result = calculate_months_to_repay_loan(loan_principal, monthly_payment)
        print_months_to_repay(result)
    elif action == 'p':
        number_of_months = int(input("Enter the number of months: "))
        rounded_payment, last_payment = calculate_monthly_payment(loan_principal, number_of_months)
        print_monthly_payment(rounded_payment, last_payment)

if __name__ == '__main__':
    main()
```

```python
from math import ceil, floor, log, pow


def calc(interest, principal=0.0, payment=0.0, periods=0.0):
    rate = interest / (12 * 100)
    if not periods:
        return ceil(log(payment / (payment - rate * principal), 1 + rate))
    if not payment:
        return principal * rate * pow(1 + rate, periods) / (pow(1 + rate, periods) - 1)
    if not principal:
        return payment / (rate * pow(1 + rate, periods) / (pow(1 + rate, periods) - 1))


choice = input('What do you want to calculate?\n'
               'type "n" - for count of months,\n'
               'type "a" - for annuity monthly payment,\n'
               'type "p" - for credit principal:\n')
if choice == 'n':
    periods_finite = calc(principal=float(input('Enter credit principal:\n')),
                          payment=float(input('Enter monthly payment:\n')),
                          interest=float(input('Enter credit interest:\n')))
    print('You need',
          f'{periods_finite} months' if periods_finite < 12
          else (f'{int(periods_finite // 12)} year' if not periods_finite % 12
                else f'{int(periods_finite // 12)} years and {ceil(periods_finite % 12)} months'),
          'to repay this credit!')
elif choice == 'a':
    payment_finite = calc(principal=float(input('Enter credit principal:\n')),
                          periods=float(input('Enter count of periods:\n')),
                          interest=float(input('Enter credit interest:\n')))
    print(f'Your annuity payment = {ceil(payment_finite)}!')
elif choice == 'p':
    principal_finite = calc(payment=float(input('Enter monthly payment:\n')),
                            periods=float(input('Enter count of periods:\n')),
                            interest=float(input('Enter credit interest:\n')))
    print(f'Your credit principal = {floor(principal_finite)}!')
```

```python
import math


class CreditCalculator:
    def __init__(self):
        self.principal = 0.0    # credit principal
        self.payment = 0.0      # monthly payment
        self.months_cnt = 0.0   # count of the period
        self.interest = 0.0     # monthly interest rate
        self.target = 'n'       # calculation target

    def setup(self):
        self.target = input('What do you want to calculate?\n'
                            'type "n" - for count of months,\n'
                            'type "a" - for annuity monthly payment,\n'
                            'type "p" - for credit principal:\n')

    def set_principal(self):
        self.principal = float(input('Enter credit principal:\n'))

    def set_payment(self):
        self.payment = float(input('Enter monthly payment:\n'))

    def set_interest(self):
        self.interest = float(input('Enter credit interest:\n')) / 1200

    def set_periods(self):
        self.months_cnt = float(input('Enter count of periods:\n'))

    def get_period(self):
        self.set_principal()
        self.set_payment()
        self.set_interest()
        self.months_cnt = math.ceil(math.log(
            self.payment / (self.payment - self.interest * self.principal),
            1 + self.interest))
        print(f'You need {self.months_to_years()} to repay the credit!')

    def get_ann_payment(self):
        self.set_principal()
        self.set_periods()
        self.set_interest()
        self.payment = self.principal * (
            (self.interest * (1 + self.interest)**self.months_cnt) /
            ((1 + self.interest)**self.months_cnt - 1)
            )
        print(f'Your annuity payment = {math.ceil(self.payment)}!')

    def get_principal(self):
        self.set_payment()
        self.set_periods()
        self.set_interest()
        self.principal = self.payment / (
            (self.interest * (1 + self.interest)**self.months_cnt) /
            ((1 + self.interest)**self.months_cnt - 1)
            )
        print(f'Your credit principal = {math.floor(self.principal)}!')

    def months_to_years(self):
        result = ''
        years = self.months_cnt // 12
        months = self.months_cnt % 12

        if years:
            result += f'{years} year'
            if years % 10 != 1 or years == 11:
                result += 's'

        if years and months:
            result += ' and '

        if months:
            result += f'{months} month'
            if months > 1:
                result += 's'
        return result

    def calculate(self):
        if self.target == 'n':
            self.get_period()
        elif self.target == 'a':
            self.get_ann_payment()
        elif self.target == 'p':
            self.get_principal()
        else:
            print('Command unknown')

    def run(self):
        self.setup()
        self.calculate()


calculator = CreditCalculator()
calculator.run()
```

```python
import math


class Parameter:
    '''Calculation parameter template.'''
    def __init__(self, value, type_, prompt):
        self.value = value
        self.type_ = type_
        self.prompt = prompt


class CreditCalculator:
    '''Credit calculator is able to work with annual type of payment.'''

    def __init__(self):
        self.main_prompt = 'What do you want to calculate?'
        self.option_prompt = (
            'type "n" - for count of months,\n'
            'type "a" - for annuity monthly payment,\n'
            'type "p" - for credit principal:\n'
        )

        self.parameters = {
            'credit_principal': Parameter(None, int, 'Enter credit principal'),
            'monthly_payment': Parameter(None, float, 'Enter monthly payment'),
            'credit_interest': Parameter(None, float, 'Enter credit interest'),
            'count_of_month': Parameter(None, int, 'Enter count of periods'),
        }

    def run(self):
        print(self.main_prompt)
        option = input(self.option_prompt)

        if option == 'n':  # count of month
            self.get_parameters('credit_principal', 'monthly_payment', 'credit_interest')
            self.print_count_of_month(self.calculate_count_of_month())

        elif option == 'a':  # annuity monthly payment
            self.get_parameters('credit_principal', 'count_of_month', 'credit_interest')
            print(f'Your annuity payment = {self.calculate_monthly_payment()}!')

        elif option == 'p':  # credit principal
            self.get_parameters('monthly_payment', 'count_of_month', 'credit_interest')
            print(f'Your credit principal = {self.calculate_credit_principal()}!')

        else:
            print(f'You enter wrong option: "{option}"')

    def get_parameters(self, *parameter_names):
        '''Get calculation parameters by theirs names from the input.'''
        for p_name in parameter_names:
            p = self.parameters[p_name]
            p.value = p.type_(input(p.prompt + ':\n'))

    def calculate_count_of_month(self):
        '''Calculate count of month for return a credit with annuity payments.'''
        p = self.parameters
        interest_rate = p['credit_interest'].value / (12 * 100)
        count_of_month = math.log(
            (
                p['monthly_payment'].value
                /  # -----------------------divide the top by the bottom-----------------
                (p['monthly_payment'].value - interest_rate * p['credit_principal'].value)
            ),
            1 + interest_rate
        )
        return math.ceil(count_of_month)  # count_of_month

    def calculate_monthly_payment(self):
        '''Calculate monthly payment for return a credit with annuity payments.'''
        p = self.parameters
        interest_rate = p['credit_interest'].value / (12 * 100)
        monthly_payment = p['credit_principal'].value * interest_rate * (1 + interest_rate) ** p['count_of_month'].value
        monthly_payment /= (1 + interest_rate) ** p['count_of_month'].value - 1

        return math.ceil(monthly_payment)

    def calculate_credit_principal(self):
        '''Calculate credit principal for return a credit with annuity payments.'''
        p = self.parameters
        interest_rate = p['credit_interest'].value / (12 * 100)
        credit_principal = p['monthly_payment'].value * ((1 + interest_rate) ** p['count_of_month'].value - 1)
        credit_principal /= interest_rate * (interest_rate + 1) ** p['count_of_month'].value

        return math.floor(credit_principal)

    @staticmethod  # we use a static method when we dont't refer to 'self'
    def print_count_of_month(count_of_month):
        '''Convert number of month to years, months form and print'''
        years, months = int(count_of_month // 12), math.ceil(count_of_month % 12)
        msg = f'You need'
        msg += '' if years == 0 else ' 1 year' if years == 1 else f' {years} years'
        msg += '' if months == 0 else ' and 1 month' if months == 1 else f' {months} months'
        msg += ' to repay this credit!'
        print(msg)


if __name__ == '__main__':

    calc = CreditCalculator()
    calc.run()
```

```python


import math

MSG_WHAT_TO_CALCULATE = "What do you want to calculate? \n" \
                        "type \"n\" for number of monthly payments, \n" \
                        "type \"a\" for annuity monthly payment amount, \n" \
                        "type \"p\" for loan principal:"

CALC_N = 'n'
CALC_A = 'a'
CALC_P = 'p'

UNIT_MONTHS = 'months'
UNIT_MONTH = 'month'
UNIT_YEARS = 'years'
UNIT_YEAR = 'year'


def ask_what_to_calculate():
    return input(MSG_WHAT_TO_CALCULATE + "\n")


def ask_p():
    return int(input('Enter the loan principal:' + "\n"))


def ask_a():
    return float(input('Enter the annuity payment:' + "\n"))


def ask_n():
    return int(input('Enter the number of periods:' + "\n"))


def ask_i():
    annual_interest = float(input('Enter the loan interest:' + "\n"))
    return annual_interest / (12 * 100)


def calc_p(a, n, i):
    return a / ((i * math.pow(1 + i, n)) / (math.pow(1 + i, n) - 1))


def calc_a(p, n, i):
    return p * (i * math.pow(1 + i, n)) / (math.pow(1 + i, n) - 1)


def calc_n(p, a, i):
    return math.ceil(math.log(a / (a - i * p), 1 + i))


def print_p(p):
    print('Your loan principal = {principal}!'.format(principal=round(p)))


def print_a(a):
    print('Your monthly payment = {payment}!'.format(payment=math.ceil(a)))


def print_n(n):
    years = int(n // 12)
    months = math.ceil(n % 12)
    year_unit = UNIT_YEARS if years > 1 else UNIT_YEAR
    month_unit = UNIT_MONTHS if months > 1 else UNIT_MONTH

    if years > 0 and months > 0:
        print("It will take {y} {y_unit} and {m} {m_unit} to repay this loan!"
              .format(y=years, m=months, y_unit=year_unit, m_unit=month_unit))
    elif years > 0:
        print("It will take {y} {y_unit} to repay this loan!".format(y=years, y_unit=year_unit))
    elif months > 0:
        print("It will take {m} {m_unit} to repay this loan!".format(m=months, m_unit=month_unit))


def main():

    what_to_calc = ask_what_to_calculate()

    if CALC_P == what_to_calc:
        a, n, i = (ask_a(), ask_n(), ask_i())
        p = calc_p(a, n, i)
        print_p(p)

    elif CALC_A == what_to_calc:
        p, n, i = (ask_p(), ask_n(), ask_i())
        a = calc_a(p, n, i)
        print_a(a)

    elif CALC_N == what_to_calc:
        p, a, i = (ask_p(), ask_a(), ask_i())
        n = calc_n(p, a, i)
        print_n(n)


if __name__ == "__main__":
    main()
```
