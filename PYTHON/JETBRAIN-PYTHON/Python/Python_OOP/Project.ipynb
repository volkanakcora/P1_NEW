You will also create a simple menu with two options: "play" and "exit". The play option lets the user input their name, after which the game greets the user by that name and asks them if they are ready to begin the adventure.

Objectives
- Choose a name for your game and display it using online generated ASCII art. Your title should take up exactly 8 non-empty lines. For this task, it will be useful to look at some examples of text art.
- Display the player's options in the main menu. In this stage, these options are Play and Exit shown in this exact order on separate lines right below the title.

- If they respond yes, the program should print a short message telling the user to get back to coding and end the program. If they choose no, keep asking the same question until they choose yes.
- If the player chooses exit on the title screen, print a message and end the program.
- If the player enters an invalid command, make sure you point it out by printing "Invalid input" and ask for a valid input again.
>Note that command words should be in brackets so that the player knows what to type. Please use the same command words that we're using in the descriptions and examples. This applies to all stages of the project.

````python
import argparse
import random
import time
import sys


class Game:
    def __init__(self, seed, min_duration, max_duration, possible_locations):
        random.seed(seed)
        self.min_duration = min_duration
        self.max_duration = max_duration
        self.possible_locations = possible_locations.split(',')

        self.total_titanium = 0

    def explore(self):
        print("Searching...")
        num_locations = min(random.randint(1, 9), len(self.possible_locations))

        if num_locations == 0:
            print("No locations to explore.")
            return

        locations_to_explore = random.sample(self.possible_locations, num_locations)

        for idx, location in enumerate(locations_to_explore, start=1):
            print(f"[{idx}] {location.replace('_', ' ')}")

            input("[S] to continue searching")

        print("Nothing more in sight.")

        while True:
            choice = input("[Back]")
            if choice.lower() == 'back':
                return
            try:
                choice = int(choice)
                if 1 <= choice <= len(locations_to_explore):
                    self.explore_location(locations_to_explore[choice - 1])
                    break
                else:
                    print("Invalid location number")
            except ValueError:
                print("Invalid command")

    def explore_location(self, location):
        print("Deploying robots...")
        random.seed(location)  # Use the location as a seed for consistent randomness
        titanium_yield = random.randint(10, 100)
        self.total_titanium += titanium_yield
        print(f"{location.replace('_', ' ')} explored successfully, with no damage taken.")
        print(f"Acquired {titanium_yield} lumps of titanium")
        while True:
            self.show_hub()

    def main_menu(self):
        while True:
            print("+=============================================================================+")
            print("     ######*   ##*   ##*  #######*  ##*  ##*  #######*  ######*   #######*")
            print("     ##*  ##*  ##*   ##*  ##*       ##* ##*   ##*       ##*  ##*  ##*      ")
            print("     ##*  ##*  ##*   ##*  #######*  #####*    #####*    ######*   #######*")
            print("     ##*  ##*  ##*   ##*       ##*  ##* ##*   ##*       ##*  ##*       ##*")
            print("     ######*    ######*   #######*  ##*  ##*  #######*  ##*  ##*  #######*")
            print("                         (Survival ASCII Strategy Game)")
            print("+=============================================================================+")
            print("\n[Play]\n[High] Scores\n[Help]\n[Exit]")

            command = input("\nYour command: ").strip().lower()

            if command == 'play':
                self.play_game()
            elif command == 'high':
                if self.total_titanium == 0:
                    print("No scores")
                else:
                    print(f"Total titanium: {self.total_titanium}")
                # Prompt user for back to game or invalid input
                while True:
                    back_command = input("\n[Back] to game: ").strip().lower()
                    if back_command == 'back':
                        break
                    else:
                        print("Invalid input")
            elif command == 'help':
                print("Coming SOON! Thanks for playing!")
                break
            elif command == 'exit':
                print("Thanks for playing, bye!")
                break
            else:
                print("Invalid input")

    def play_game(self):
        name = input("\nEnter your name: ")
        print(f"\nGreetings, commander {name}!")
        start = input("Are you ready to begin?\n[Yes] [No] Return to Main[Menu]\n\nYour command: ").strip().lower()

        if start == 'yes':
            self.show_hub()

    def show_hub(self):
        print("+==============================================================================+")
        print("  $   $$$$$$$   $  |  $   $$$$$$$   $  |  $   $$$$$$$   $")
        print("  $$$$$     $$$$$  |  $$$$$     $$$$$  |  $$$$$     $$$$$")
        print("      $$$$$$$      |      $$$$$$$      |      $$$$$$$")
        print("     $$$   $$$     |     $$$   $$$     |     $$$   $$$")
        print("     $       $     |     $       $     |     $       $")
        print("+==============================================================================+")
        print("| Titanium:", self.total_titanium, " " * (76 - len(str(self.total_titanium))), "|")
        print("+==============================================================================+")
        print("|                  [Ex]plore                          [Up]grade                |")
        print("|                  [Save]                             [M]enu                   |")
        print("+==============================================================================+")

        while True:
            command = input("\nYour command: ").strip().lower()

            if command == 'ex':
                self.explore()
            elif command == 'exit':
                print("Thanks for playing, bye!")
                break
            elif command == 'm':
                self.print_second_menu()
            else:
                print("Invalid command")

    def print_second_menu(self):
        menu = """
        |==========================|
        |            MENU          |
        |                          |
        | [Back] to game           |
        | Return to [Main] Menu    |
        | [Save] and exit          |
        | [Exit] game              |
        |==========================|"""
        print(menu)
        command = input("\nYour command: ").strip().lower()
        while True:
            if command == 'back':
                self.show_hub()
            elif command == 'main':
                self.main_menu()
            elif command == 'save':
                print("Coming SOON! Thanks for playing!")
                sys.exit()
            elif command == 'exit':
                print("Thanks for playing, bye!")
                sys.exit()
            else:
                print("Invalid input")


if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Survival ASCII Strategy Game")
    parser.add_argument("Seed", help="Seed for pseudo-random number generation")
    parser.add_argument("MinimumDuration", type=int, default=0, nargs='?', help="Minimum duration of animations")
    parser.add_argument("MaximumDuration", type=int, default=0, nargs='?', help="Maximum duration of animations")
    parser.add_argument("PossibleLocations", nargs='?', default="High_street,Green_park,Destroyed_Arch",
                        help="Names of possible locations separated by comma")

    args = parser.parse_args()

    game = Game(args.Seed, args.MinimumDuration, args.MaximumDuration, args.PossibleLocations)
    game.main_menu()
````
