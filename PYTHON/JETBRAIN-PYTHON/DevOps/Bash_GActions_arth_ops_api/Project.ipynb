```python
#!/bin/bash

score=0

while true; do
    # Display the main menu
    echo "Welcome to the True or False Game!"
    echo "0. Exit"
    echo "1. Play a game"
    echo "2. Display scores"
    echo "3. Reset scores"
    echo "Enter an option:"

    # Read the option from the user
    read option

    # Check the selected option
    case $option in
        0)
            echo "See you later!"
            exit ;;
        1)
            # Ask for the player's name
            echo "What is your name?"
            read player_name

            # Initialize game variables
            correct_answers=0
            seed=4096
            responses=("Perfect!" "Awesome!" "You are a genius!" "Wow!" "Wonderful.")

            # Start the game loop
            while true; do
                # Connect to the /game endpoint and retrieve question and answer
                question_answer=$(curl -s http://127.0.0.1:8000/game)
                question=$(echo "$question_answer" | sed 's/.*"question": *"\{0,1\}\([^,"]*\)"\{0,1\}.*/\1/')
                answer=$(echo "$question_answer" | sed 's/.*"answer": *"\{0,1\}\([^,"]*\)"\{0,1\}.*/\1/')

                # Print the question to the standard output
                echo "Question: $question"

                # Ask if it is True or False
                echo "True or False?"
                read user_response

                # Compare user response with the answer
                if [ "$user_response" == "$answer" ]; then
                    # Correct answer
                    echo "${responses[$((RANDOM % 5))]}"
                    correct_answers=$((correct_answers + 1))
                    score=$((score + 10))
                else
                    # Wrong answer
                    echo "Wrong answer, sorry!"
                    echo "$player_name, you have $correct_answers correct answer(s)."
                    echo "Your score is $score points."

                    # Save the player's score to scores.txt
                    echo "User: $player_name, Score: $score, Date: $(date +%Y-%m-%d)" >> scores.txt
                    break
                fi
            done ;;
        2)
            # Display scores
            if [ -e "scores.txt" ]; then
                echo "Player scores:"
                cat scores.txt
            else
                echo "File not found or no scores in it!"
            fi ;;
        3)
            # Reset scores
            if [ -e "scores.txt" ]; then
                rm scores.txt
                echo "File deleted successfully!"
            else
                echo "File not found or no scores in it!"
            fi ;;
        *)
            echo "Invalid option!" ;;
    esac
done
```
